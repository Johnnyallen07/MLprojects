## Additional Feature Engineering

The accuracy is 50% only after submission no matter what model is chosen, hence, it needs some improvement

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = 'data/train.csv'
df = pd.read_csv(path)
```

Passenger_Id has form aaaa_bb, it might be a clue that a group the passenger is traveling together, and bb is their id inside the group

```python
df['Group'] = [x.split('_')[0] for x in list(df['PassengerId'])]
df['Group'] = df['Group'].astype(int)

# New feature - Group size
df['Group_size']=df['Group'].map(lambda x: df['Group'].value_counts()[x])

plt.figure()
sns.countplot(data=df, x='Group_size', hue='Transported')
plt.title('Group size')
```
